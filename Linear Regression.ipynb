```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv(
    "Datasets/Salary_dataset_linear_regression.csv"
)
data.head()
```

```python
X_train = data['YearsExperience']
y_train = data['Salary']
```

```python
w, b = 100, 10
m = X_train.shape[0]
m
```

```python
def costs(w, b, m, X_train, y_train):
    cost, var1, var2 = 0, 0, 0
    for i in range(m):
        z = w * X_train[i] + b
        cost += (z - y_train[i]) ** 2
        var1 += (z - y_train[i]) * X_train[i]
        var2 += (z - y_train[i])
    cost = 1 / 2 / m * cost
    var1 = 1 / m * var1
    var2 = 1 / m * var2
    return cost, var1, var2
```

```python
def gradientDescent(w, b, m,  X_train, y_train):
    alpha = 0.01
    iterations = 5000
    wn = w
    bn = b
    for i in range(iterations):
        cost, var1, var2 = costs(
            wn, bn, m,  X_train, y_train
        )
        wn = wn - alpha * var1
        bn = bn - alpha * var2
        print(f"Iteration {i}: cost = {cost}")
    return wn, bn
```

```python
wn, bn = gradientDescent(w, b, m,  X_train, y_train)
y_pred = wn * X_train + bn 
```

```python
plt.plot(X_train, y_train)
plt.plot(X_train, y_pred)
plt.show()
```
